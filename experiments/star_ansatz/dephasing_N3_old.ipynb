```python
import numpy as np
import sys
sys.path.insert(1,'../')

import matplotlib.pyplot as plt
```

```python
import csv
from numpy import *
f2 = open('../../experiments/star_ansatz/dephasing_N3.csv', 'r') 
lines = f2.readlines()

noise, cost, cls, sld, traceoff = [], [], [], [], []

for line in lines:
    p = line.split()
    noise.append(float(p[0]))
    cost.append(float(p[1]))
    cls.append(float(p[2]))
    sld.append(float(p[3]))
    traceoff.append(float(p[4]))
    
f2.close() 

fig,ax = plt.subplots()
ax.plot(noise,cost, '--o', label='cost function')
ax.plot(noise,cls, '--^',label='classical bound')
ax.plot(noise,sld, '--v',label='quantum bound')

ax2=ax.twinx()
ax2.plot(noise,traceoff, '--s', color ='red',label='trade off')
ax2.set_ylim(0,3)
ax2.spines['right'].set_color('red')
ax2.tick_params(axis='y', colors='red') 

ax.legend(loc='center')
ax2.legend()

plt.savefig('fig_noise.eps')
plt.xticks(np.arange(0,1.0,0.1))
plt.show()
```
