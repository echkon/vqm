```python
import numpy as np
import sys
sys.path.insert(1,'../')

import matplotlib.pyplot as plt
```

```python
NL = np.linspace(1,10,10)
NN = [3,4,5]
SQL = 1.0/NL #standard quantum limit
HL = 1.0/NL**2 #Heisenberg limit

cls_Mar = [0.473548385,0.282566222,0.232284449]
quan_Mar = [0.447941907,0.206571139,0.149578768]

cls_nonMar = [0.377026008,0.315494,0.184303548]
quan_nonMar = [0.302017863,0.151200834,0.059844543]


fig,ax = plt.subplots()

ax.plot(NN,cls_Mar, '-^',label='classical Mar')
ax.plot(NN,quan_Mar, '-v',label='quantum Mar')

ax.plot(NN,cls_nonMar, '--^',markerfacecolor='none',label='classical nonMar')
ax.plot(NN,quan_nonMar, '--v',markerfacecolor='none',label='quantum nonMar')

ax.plot(NL,SQL, '-',label='SQL')
ax.plot(NL,HL, '-',label='HL')

#plt.xticks(np.arange(1,10,1))
#ax.set_xticks(NN,labels=['G1', 'G2', 'G3'])

ax.legend(loc='lower left')
ax.set_xscale('log')
ax.set_yscale('log')


plt.show()
```
